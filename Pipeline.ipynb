```python
df = pd.read_csv(r"C:\Users\Heena Baugwala\OneDrive\Desktop\Job Applications\Data sets\Pipeline Project\ai_jobs.csv.csv")
print(f"Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")
df.head()
```

```python
print("Column names:")
print(df.columns.tolist())
print(f"\nData types:")
print(df.dtypes)
print(f"\nMissing values:")
print(df.isnull().sum())
```

```python
# Step 1: Remove any duplicate rows
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]
print(f"Duplicates removed: {before - after}")
print(f"Rows remaining: {after}")

# Step 2: Standardise text columns (strip whitespace, consistent case)
text_columns = ['job_title', 'job_category', 'experience_level', 'country', 'remote_work', 'company_size', 'industry']
for col in text_columns:
    df[col] = df[col].str.strip()

print("Text columns cleaned and standardised!")
```

```python
df['salary_range'] = df['salary_max_usd'] - df['salary_min_usd']
print("New column: salary_range")
print(df[['job_title', 'salary_min_usd', 'salary_max_usd', 'salary_range']].head())
```

```python
df['salary_band'] = pd.cut(df['annual_salary_usd'], 
                           bins=[0, 100000, 150000, 200000, 250000, 500000],
                           labels=['Under 100K', '100K-150K', '150K-200K', '200K-250K', '250K+'])

df['experience_rank'] = df['experience_level'].map({
    'Entry (0-2 yrs)': 1,
    'Mid (3-5 yrs)': 2,
    'Senior (6-9 yrs)': 3,
    'Lead (10+ yrs)': 4
})

print("Salary band distribution:")
print(df['salary_band'].value_counts())
print(f"\nExperience rank sample:")
print(df[['experience_level', 'experience_rank']].head(10))
```

```python
errors = 0

# Check 1: Salary min should never be higher than salary max
bad_salary = df[df['salary_min_usd'] > df['salary_max_usd']]
print(f"Check 1 — Min salary > Max salary: {len(bad_salary)} errors")
errors += len(bad_salary)

# Check 2: Annual salary should be between min and max
bad_annual = df[(df['annual_salary_usd'] < df['salary_min_usd']) | (df['annual_salary_usd'] > df['salary_max_usd'])]
print(f"Check 2 — Annual salary outside min-max range: {len(bad_annual)} errors")
errors += len(bad_annual)

# Check 3: Years of experience should not be negative
bad_exp = df[df['years_of_experience'] < 0]
print(f"Check 3 — Negative years of experience: {len(bad_exp)} errors")
errors += len(bad_exp)

# Check 4: Demand score should be between 0 and 10
bad_demand = df[(df['demand_score'] < 0) | (df['demand_score'] > 10)]
print(f"Check 4 — Demand score outside 0-10: {len(bad_demand)} errors")
errors += len(bad_demand)

print(f"\n✅ TOTAL VALIDATION ERRORS: {errors}")
if errors == 0:
    print("Data quality: PASSED — ready for export!")
else:
    print(f"Data quality: FAILED — {errors} issues need fixing")
```
